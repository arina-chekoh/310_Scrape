```python
import pandas as pd
Blue = pd.read_csv('Comment_Blue_Buffalo.csv')
Purina = pd.read_csv('Comment_Purina_Pro.csv')
```

```python
Blue
```

```python
df1 = pd.DataFrame(Blue)
df2 = pd.DataFrame(Purina)
```

```python
df1 = df1.dropna()
```

```python
df1 = df1.drop_duplicates(subset=['Detail'])
df1 = df1.rename(columns = {'Raiting' : 'Rating'})
df1
```

```python
df2 = df2.drop_duplicates(subset=['Detail'])
df2
```

```python
df2 = df2.dropna()
```

```python
data = pd.concat([df1, df2], axis=0)
```

```python
df = pd.DataFrame(data)
df
```

```python
df['Rating'] = df['Rating'].str.replace("out of 5 stars", "")
del df[df.columns[0]]
```

```python
df['Rating'] = df['Rating'].astype(float)
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.to_csv('Dog_comment_merge.csv', index=False)
```
