```python
import pandas as pd
import bs4
import requests
import openpyxl
import csv
from datetime import datetime

page = 1
title_list = []
detail_list = []
date_list = []
retrieved_date  = []

while page <= 8:
    data = requests.get(f'https://thestandard.co/page/{str(page)}/?s=%E0%B8%AD%E0%B8%B2%E0%B8%AB%E0%B8%B2%E0%B8%A3%E0%B8%AA%E0%B8%B1%E0%B8%95%E0%B8%A7%E0%B9%8C%E0%B9%80%E0%B8%A5%E0%B8%B5%E0%B9%89%E0%B8%A2%E0%B8%87&search')
    soup = bs4.BeautifulSoup(data.text)
    for n in soup.find_all('div',{'class':'news-item'}):
        title_list.append(n.find('h3' ,{'class' : 'news-title'}).find('a').text)
        detail_list.append(n.find('div',{'class' : 'desc'}).text)
        date_list.append(n.find('div',{'class' : 'date'}).text) 
        retrieved_date.append(datetime.now().strftime("%Y-%m-%d"))
    page += 1
    print('complete page number: ',page)
    
    
    
table = pd.DataFrame([title_list, detail_list, date_list, retrieved_date]).transpose()
table.columns = ['title', 'detail', 'date', 'retrieved_date']
table.set_index('title')
```

```python
table =table.set_index('title')
```

```python
table.to_csv('Pet food News.csv',encoding='utf-8-sig')
```
